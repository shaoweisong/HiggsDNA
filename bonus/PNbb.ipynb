```python
import numpy
import hist
import correctionlib
import correctionlib.convert
import numpy as np
```

```python
xedge = [100,450,600,2000]
sfs16pre=[0.932,0.782,1.042]
sfs16preup=list(np.array([0.135,0.35,0.465])+np.array(sfs16pre))
sfs16predown=list(np.array([-0.142,-0.361,-0.390])+np.array(sfs16pre))
sfhist16pre = hist.Hist(
    hist.axis.Variable(xedge, label='pt'),  # xaxis:eta
    storage=hist.storage.Double(),
    data=sfs16pre
)
sfhist16preup = hist.Hist(
    hist.axis.Variable(xedge, label='pt'),  # xaxis:eta
    storage=hist.storage.Double(),
    data=sfs16preup
)
sfhist16predown = hist.Hist(
    hist.axis.Variable(xedge, label='pt'),  # xaxis:eta
    storage=hist.storage.Double(),
    data=sfs16predown
)
sfhist16pre.name = "PNbbSF16pre"
sfhist16pre.label = "out"
PNetbbsfs16pre = correctionlib.convert.from_histogram(sfhist16pre)
PNetbbsfs16pre.description = "scale factors for ParticleNet Xbb tagger 2016preVFP"
PNetbbsfs16pre.data.flow = "clamp"

sfhist16preup.name = "PNbbSF16preup"
sfhist16preup.label = "out"
PNetbbsfs16preup = correctionlib.convert.from_histogram(sfhist16preup)
PNetbbsfs16preup.description = "scale factors for ParticleNet Xbb tagger 2016preVFP"
PNetbbsfs16preup.data.flow = "clamp"

sfhist16predown.name = "PNbbSF16predown"
sfhist16predown.label = "out"
PNetbbsfs16predown = correctionlib.convert.from_histogram(sfhist16predown)
PNetbbsfs16predown.description = "scale factors for ParticleNet Xbb tagger 2016preVFP"
PNetbbsfs16predown.data.flow = "clamp"
```
